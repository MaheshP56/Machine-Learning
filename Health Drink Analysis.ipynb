```python
#here we are seperating object and numerical data types
obj_col=[]
num_col=[]
for col in bn0.columns:
    if bn0[col].dtype=='O':
        obj_col.append(col)
    else:
        num_col.append(col)
```

```python
print("Object data type features ",obj_col)
print("Numerical data type features ",num_col)
```
